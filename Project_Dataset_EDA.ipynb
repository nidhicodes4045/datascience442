```python
from scipy import stats

def find_non_normal_columns(df, alpha=0.05):
    """
    Checks each column in a DataFrame for normality using the Shapiro-Wilk test.

    Args:
        df (pd.DataFrame): The DataFrame to check.
        alpha (float, optional): The significance level for the test. Defaults to 0.05.

    Returns:
        list: A list of column names that are not normally distributed.
    """
    non_normal_columns = []
    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):  # Check if the column is numeric
            try:
                stat, p = stats.shapiro(df[column])
                if p < alpha:
                    non_normal_columns.append(column)
            except ValueError:
                print(f"Warning: Not enough data points in '{column}' for Shapiro-Wilk test.")
        else:
            print(f"Warning: '{column}' is not numeric, skipping normality test.")

    return non_normal_columns
#find_non_normal_columns(df)
```

```python
numerical_df = df.copy()
numerical_df.drop(['County', 'State', 'mergedkeycol'], axis = 1, inplace = True)

def plot_histograms_by_percentile(df, percentile):
    for column in df.columns:
        percentile_95 = np.percentile(df[column], percentile)
        plt.figure()
        sns.histplot(df[column])
        plt.xlim(0, right=percentile_95)
        plt.title(f'Histogram of {column} with {percentile}th Percentile Limit')
        plt.show()

#plot_histograms_by_percentile(numerical_df, 95)
```

```python
#demographic_df = df.copy()
#demographic_df = demographic_df[['Age.Percent 65 and Older', 'Age.Percent Under 18 Years', 'Age.Percent Under 5 Years', '']]

def plot_histograms_by_IQR(df):
    for column in df.columns:
      Q1 = df[column].quantile(0.25)
      Q3 = df[column].quantile(0.75)
      IQR = Q3 - Q1
      upper_limit = Q3 + 1.5 * IQR  # Adjust the multiplier as needed
      plt.figure()
      sns.histplot(df[column])
      plt.xlim(0, right=upper_limit)
      plt.title('Histogram with IQR Limit')
      plt.show()
#plot_histograms_by_IQR(numerical_df)
```

## check skewness for all columns,

skewness is between 0 and 1


```python
right_skewed = []
left_skewed = []
relative = []
def check_skewness(df):
      for column in df.columns:
          skewness = df[column].skew().round(2)
          if skewness >= 1:
              right_skewed.append(column)
              print(f"'{column}' is right-skewed (skewness: {skewness})")
          elif skewness <= -1:
              left_skewed.append(column)
              print(f"'{column}' is left-skewed (skewness: {skewness})")
          else:
              relative.append(column)
              print(f"'{column}' is relatively symmetric (skewness: {skewness})")

check_skewness(numerical_df)
print("\nRight-skewed columns:", right_skewed)
print("Left-skewed columns:", left_skewed)
print("Relatively symmetric columns:", relative)
```

## Bivariate Analysis

---



*   Explore the relationship between two variables



```python
def plot_scatter_matrix(df):

    for x_col in df.columns:
      for y_col in df.columns:
        if x_col != y_col:
          plt.figure() #create a new figure for each plot.
          df.plot(kind='scatter', x=x_col, y=y_col, title=f'{y_col} vs {x_col}')
          plt.show()
#plot_scatter_matrix(numerical_df)
```
